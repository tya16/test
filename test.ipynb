```python
# Importing required library
import numpy as np
import matplotlib.pyplot as plt

# Creating x axis with range and y axis
# Function for Plotting Sine and Cosine Graph
x = np.arange(0, 5*np.pi, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

# Plotting Sine and Cosine Graphs
plt.plot(x, y1, color='green', label='Sine')
plt.plot(x, y2, color='darkblue', label='Cosine')
plt.title("Sine and Cosine Graphs")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.legend()

# Saving the plot as a PNG image
plt.savefig('sine_cosine_plot.png', format='png', dpi=300)  # Simpan dengan resolusi 300 DPI

# Display the plot
plt.show()
```
